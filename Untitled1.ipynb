```python
%%time

Nsat = 100
Nhalo = 1000

halos = np.ndarray((1000, 3, 100))

for h in range(Nhalo):

    r = samp.rejection(p, Nsat, xrange=(0, 5), yrange=(0, 2)).T
    angles = samp.spherical(Nsat).T
    theta, phi = angles[0], angles[1]

    galaxies = np.array([r[0], theta, phi])
    
    halos[h, :, :] = galaxies

halos.shape
```

```python
nbins = 20

bins = np.log10(np.logspace(-4, np.log10(5), nbins+1))
halo_hist = np.ndarray((1000, nbins))

for h, halo in enumerate(halos):
    halo_hist[h, :] = np.histogram(np.log10(halo[0]), bins)[0]
    
halo_hist.shape
```

```python
N = lambda x: dist.n(x) * 4*np.pi*x**2

plt.plot(np.log10(truex), N(truex), lw=5, alpha=.5, label='true')


halo_hist_mean = rt.mean(halo_hist)

plt.bar(bins[:-1], halo_hist_mean, .2, alpha=.5, color='C1')

plt.yscale('log')
plt.ylim([1e-3, 1e6])

plt.show()
```

```python
from scipy.optimize import brent

N = lambda x: dist.n(x) * 4*np.pi*x**2
Nmaxx = brent(lambda x: -N(x), brack=(1e-4, 5))
Nmaxy = N(Nmaxx)

plt.plot(truex, N(truex), lw=5, alpha=.5, label='true')
plt.axvline(Nmaxx, c='k', linestyle='--', lw=.5)
plt.axhline(Nmaxy/2, c='k', linestyle='--', lw=.5)

plt.show()
```

```python
Nshift = lambda x: N(x)-Nmaxy/2

plt.plot(truex, N(truex), lw=5, alpha=.5, label='true')
plt.axhline(Nmaxy/2, c='k', linestyle='--', lw=.5)

r1 = rf.bisect(Nshift, 1e-4, Nmaxx)
r2 = rf.bisect(Nshift, Nmaxx, 5)

plt.axvline(r1, c='k', linestyle='--', lw=.5)
plt.axvline(r2, c='k', linestyle='--', lw=.5)
plt.show()

np.round((r1, r2), 3)
```

```python
def median(values):
    return percentile(values, .5)
    
def percentile(values, k):
    index = len(values)*k
    if index % 1 == 0.0:     # if not a whole number
        index = int(index)+1 # round up to whole number
        return values[index]

    return rt.mean([values[index], values[index+1]])
```
