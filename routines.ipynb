```python
import numpy as np
```

```python
def prod(array):
    total = 1
    for val in array:
        total *= val
        
    return total
```

```python
def factorial(n):
    return prod(range(1, n+1))
```

```python
def poisson(l, k):
    if l < 0:
        raise Exception('lambda cannot be negative')
        
    return (l**k * np.exp(-l)) / factorial(k)
```

```python
import matplotlib.pyplot as plt

for mu in np.linspace(1, 5, 4):
    x = range(15)
    y = [poisson(mu, k) for k in x]

    plt.plot(x, y)

plt.show()
```
